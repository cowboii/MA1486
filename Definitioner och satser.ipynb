__Ex:__

För alla reella tal $a$ och $b$ definieras räkneregeln * genom

$$ a * b = a \cdot b - 1$$

där $a \cdot b = ab $ är vanligt multiplikation mellan $a$ och $b$. Undersök om
den _associativa lagen_

$$ a * (b * c) = (a * b) * c$$

gäller för dessa regler.

__Lösning:__

Vänsterledet:

$$ a * (b * c) = a * (bc - 1) - 1 = a(bc -1) - 1 = abc - a - 1 $$ 

Högerledet:

$$ (a * b) * c = (ab - 1) * c = (ab - 1)c - 1 = abc - c - 1 $$

För att associativa lagen ska gälla ska lagen gälla för alla möjliga tal $a, b$
och $c$. Ovanstående likhet gäller bara då $a = c$, så räkneregeln uppfyller
inte den associativa lagen.

### Sats, hypotes och axiom

En _sats_ är en utsaga som är sann. Att usagan är sannn verifieras med ett bevis.

En _hypotes_ är en preliminär sats, förstadiet till en sats. en hypotes är
en utsaga som antas vara sann. Hypotesen övergår till sats då/om den kan visas vara sann.

En _axiom_ är en utsaga som antas vara sann utan bevis. Axiom är till skillnad från definitioner
en grund för existensen av ett matematisk område. (Axiomatisk talsystem, på slutet av kompendiet)

Exempel på axiom är _välordningsprincipen_: bland varje uppsättning icke-negativa heltal
så finns det ett minsta tal.

(Banach-Tarskis paradox)

### Bevis

Ett bevis är en strukturerad och noggrann argumentation om varför 
påståendet/påståendena i en sats är sant/sanna.
Den logiska argumentationen ska sakna luckor och stödja sig på 
kända satser, egna slutsatser, definitioner och axiom.

Satslogik är användbar!

Olika typer av bevis (finns flera)

- Direkt bevis
- Indirekt bevis
- Motsägelsebevis
- Induktionsbevis (Vilar inte enbart på logik utan {MORE}

__Direkt bevis:__

Visa att om $n$ är ett udda tal så är $n^2$ ett udda tal.

Metod:
$P$: "$n$ är ett udda tal"<br>
$Q$: "$n^2$ är ett udda tal"

Bevisa $P \Rightarrow Q$ genom att anta att $P$ är sant och visa att det då följer att $Q$ är sant.

$P$ | $Q$ | $P \Rightarrow Q$
:-:|:-:|:-:
s | s | s
s | f | f
f | s | s
f | f | s

Antag att $P$ är sant, dvs att $n$ är udda. Då finns det ett heltal $k$ sådant att 

$$n = 2k + 1$$

Då är 

\begin{align}
n^2 = (2k+1)^2 &= 4k^2 + 4k + 1 \\
&= 2(2k^2 + 2k) + 1 
\end{align}

som är udda eftersom $2k^2 + 2k$ är ett heltal. Vi har visat att $Q$ är sant.

__Indirekt bevis:__

Låt $n$ beteckna ett heltal. Visa at om $n^2$ är ett udda tal så är $n$ också ett udda tal.

Metod:

$P$: "$n^2$ är ett udda tal"<br>
$Q$: "$n$ är ett udda tal"

Vi villa visa $P \Rightarrow Q$

$$n^2 = 2k + 1 \\
n = \sqrt{n^2} = \sqrt{2k+1}$$

Väljer att visa $\neg Q \Rightarrow \neg P$

Dessa två implikationer är logiska ekvivalenta.

$P$ | $Q$ | $P \Rightarrow Q$ |$\neg Q \Rightarrow \neg P$
:-:|:-:|:-:|:-:
s | s | s | f __s__ f
s | f | f | s __f__ f
f | s | s | f __s__ s
f | f | s | s __s__ s

$ \neg Q$: "$n$ är ett jämnt tal"<br>
$ \neg P$: "$n^2$ är ett jämnt tal"

__Ex:__

Visa att om $n^2$ är jämnt delbart med 3, så är även $n$ det.

Metod: 
$P$: "$n^2$ är jämnt delbart med 3"<br>
$Q$: "$n$ är jämnt delbart med 3"

Vi vill visa $P \Rightarrow Q$. Väljer att visa $\neg Q \Rightarrow \neg P$.

$ \neg Q$: "$n$ är inte jämnt delbart med 3"<br>
$ \neg P$: "$n^2$ är inte jämnt delbart med 3"<br>
$ \neg \Rightarrow \neg P$: "om $n$ är inte jämnt delbart med 3,
så är $n^2$ inte jämnt delbart med 3"

\end{align}

__Motsägelsebevis:__

Låt $n$ vara ett heltal. Visa att om $n^3 + 3n^2 - n + 1$ är 
ett jämnt tal, så är $n^3 + 3n^2 + n$ ett udda tal.

Vi vill visa att $P \Rightarrow Q$ är sann.
Väljer att visa $\neg(P \wedge Q)$ är sann.

Även dessa två påståenden är logiskt ekvivalenta. (Visa det!)
I ett motsägelsebevis antar man att $ P \wedge \neg Q$ är sann och visar att det leder till en motsägelse.
