# Problemlösning

Ett problem är en uppgift för vilken det gäller att main inte har någon standardmetod för att hitta lösningen.

Ett uppgift som när man läser den från början har inte en klar lösning/metod som kan användas för att lösa den.

Problem kan uppfattas olika beroande på individens ernfarenhet.

### Vad är en lösningen till ett problem?

En lösning är en redogörlse i textform, där hela lösningsprocessen beksrivs i detalj

__Ex:__

\begin{align}
    \text{Summan av alla multiplikationer mellan ettan och nian.}
\end{align}

### Fördelar med noggran redogörelse

- Andra personer kan lättare läsa och förstå din lösning. Ta hänsyn till läsare och anpassa sig till den.
- Det är lättare att sätta sig in i problemet igen om man har haft utförliga lösningar på varje litet delsteg.
- Under en noggrann redogörelse upptcker man eventuall brister.
- När man tvingar sig själv att med egna ord förklara lösningen inser man lättare ur och varför en metod man använt fungerar och vilka förutsättningar som måste vara uppfylda.
- Det ger en överblick över lösningen och därmed mölighete att avgöre om lösninge ör och om det finns en mer elegant lösning.
