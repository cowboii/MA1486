Def. _sanningsvärden_ : _Sant/Falsk_

En _utsaga_ eller _påstående_ har formen att den är antingen sann eller falsk.

__Ex:__  Ej korrekta påstående enligt ovan: 

"Denna mening är falsk." – Barber paradox.

__Ex:__ Korrekta påståenden:

<p style="color: red">"Månen är gjord av ost."</p>
<p style="color: green"> 1 + 1 = 2</p>
<p style="color: red"> 1 > 3</p>
<p style="color: green">Om $ x > 1 $, så $ x^2 > 1 $</p>
<p style="color: red">Om $ x \ge 1 $, så $ x^2 > 1 $</p>

__Ex:__ Påståenden med okännd sanningsvärde:

$ x^2 = 4 $

<hr/>

I fortsättningen betecknar: $P,Q,R,S,\dots$ eller $P_1,P_2,\dots,R_1,R_2,\dots$ olika påståenden.

__Ex:__

$ P: 1 + 1 = 2 $ "Påstående att ett plus ett är lika med två." _Bevis finns i slutet av kompendiet_

$ Q: $ Vinkelsumma i en triangel är $180^\circ$

$ P_1(x): x > 1$

$ P_2(x): x^2 > 1$

<p style="color: green">$ P_1(3)$ är sant </p>

<p style="color: red">$ P_1(0)$ är falsk</p>

$ P_3(x): $Om $P_1(x)$, så $P_2(x)$

$ R(x,y): $ Om $ x<y$, så finns det ett heltal $n$ sådant att $nx>y$.

### Negation $\neg$ (NOT)

$\neg P$ är motsatsen till $P$

__Ex:__

$ P: 1+1=2 $

$\neg P: 1+1\neq2$

$Q: x > 2019 $

$\neg Q: x \le 2019 $


$\neg Q: x < 2019$ eller $ x=2019 $ Låt $A$ vara en mängd 

$$\complement A \{ x \in \mathbf{U}: x \in A \} = 
\{ x \in \mathbf{U}: \neg (x \in A) \} $$

### Konjuktion $\wedge$ (AND)

$ P \wedge Q $ utläses "$P$ och $Q$"

Ett nytt påstående som är sant då både $P$ och $Q$ är sanna.

 P | Q ||| P ^ Q 
:-:|:-:|||:-----:
 S | S |||   S   
 F | S |||   F   
 S | F |||   F   
 F | F |||   F   
 
__Ex:__
 
$$
P(x): x > 1 \\
Q(x): x < 0 \\
P(x) \wedge Q(x): x>1 ~\text{och}~ x<0 = Falsk
$$
 
__Ex:__

\begin{align}
&P: 1 + 1 = 2 \\
&Q: ~\text{Robert gillar matte} \\
&P \wedge Q ~\text{Alltid sant}
\end{align}

__Ex:__

$$ A \cap B = \{ x \in \mathbf{U}: x \in A \wedge x \in B \} $$

### Disjunktion $\vee$ (XOR)

$ P \vee Q $ uläses "$P$ eller $Q$" är sann då minst en av
$P$ eller $Q$ är sann, annars falsk.

P | Q | P v Q
:-:|:-:|:-:
S | S | S
S | F | S
F | S | S
F | S | F

__Ex:__

$$ P(x) \vee Q(x): x > 1 ~\text{eller}~ x < 0 $$

$$ x > 0 ~\text{eller}~ x < 1 $$

__Ex:__

$$ A \cup B = \{ x \in \mathbf{U}: x \in A \vee x \in B \} $$

### Ekvivalens $\Leftrightarrow$ (XNOR)

$P \Leftrightarrow Q$ utläses "$P$ om och endast om $Q$" är sann då $P$ och $Q$ har samma sanningsvärde.

P | Q | P <=> Q 
:-:|:-:|:-:|
S | S | S
S | F | F
F | S | F
F | F | S

__Ex:__

$$ 3x+5 =1 \Leftrightarrow 3x = -4 \Leftrightarrow x = -\frac{4}{3} $$
